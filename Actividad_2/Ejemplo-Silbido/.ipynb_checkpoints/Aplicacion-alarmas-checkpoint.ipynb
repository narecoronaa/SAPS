```python
def analizar_audio(filename):
    
    fs, data = wavfile.read(filename) # frecuencia de muestreo y datos de la señal

    # Definición de parámetro temporales
    ts = 1 / fs                     # tiempo de muestreo
    N = len(data)                   # número de muestras en el archivo de audio
    t = np.linspace(0, N * ts, N)   # vector de tiempo
    senial = data#[:, 1]            # se extrae un canal de la pista de audio (si el audio es estereo)
    senial = senial * 3.3 / (2 ** 16 - 1) # se escala la señal a voltios (considerando un CAD de 16bits y Vref 3.3V)
    
    
    #################### Cálculo de la Transformada de Fourier ###################
    
    freq = fft.fftfreq(N, d=1/fs)   # se genera el vector de frecuencias
    senial_fft = fft.fft(senial)    # se calcula la transformada rápida de Fourier
    
    # El espectro es simétrico, nos quedamos solo con el semieje positivo
    f = freq[np.where(freq >= 0)]      
    senial_fft = senial_fft[np.where(freq >= 0)]
    
    # Se calcula la magnitud del espectro
    senial_fft_mod = np.abs(senial_fft) / N     # Respetando la relación de Parseval
    # Al haberse descartado la mitad del espectro, para conservar la energía 
    # original de la señal, se debe multiplicar la mitad restante por dos (excepto
    # en 0 y fm/2)
    senial_fft_mod[1:len(senial_fft_mod-1)] = 2 * senial_fft_mod[1:len(senial_fft_mod-1)]
    if len(senial_fft_mod) > 2:
        senial_fft_mod[1:-1] = 2 * senial_fft_mod[1:-1]
    
    return t, senial, f, senial_fft_mod
```

```python
# Archivos a analizar
archivos = ['alarmas/alarm.wav','alarmas/cooper_fulleon_sounder_tone_3.wav','alarmas/cooper_fulleon_sounder_tone_6.wav','alarmas/cooper_fulleon_sounder_tone_9.wav']

fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(12, 14))  # 4 filas, 2 columnas

for i, archivo in enumerate(archivos):
    t, senial, f, senial_fft_mod = analizar_audio(archivo)
    
    # Señal temporal (columna 0)
    axs[i, 0].plot(t, senial)
    axs[i, 0].set_xlabel('Tiempo [s]')
    axs[i, 0].set_ylabel('Tensión [V]')
    axs[i, 0].set_title(f'Señal temporal: {archivo}')
    axs[i, 0].grid()

    # Espectro de magnitud (columna 1)
    axs[i, 1].plot(f, senial_fft_mod)
    axs[i, 1].set_xlabel("Frecuencia [Hz]")
    axs[i, 1].set_ylabel("Magnitud")
    axs[i, 1].set_title(f'Espectro de Magnitud: {archivo}')
    axs[i, 1].grid()

plt.tight_layout()
plt.show()
```

## Selección de alarma y justificación

De acuerdo al análisis espectral y temporal de las cuatro alarmas, se selecciona **cooper_fulleon_sounder_tone_9.wav** como la alarma más adecuada para la aplicación, gracias a su espectro claro y frecuencias óptimas.

**Criterio de selección:**

- **Claridad espectral:** Su espectro de magnitud muestra dos picos muy definidos en torno a los 2000 Hz y 3000 Hz, lo que facilita la detección automática y reduce el riesgo de confusión con ruidos de ambiente.
- **Frecuencia dominante:** Las frecuencias principales están en el rango óptimo para la detección, lejos del ruido habitual y de la voz humana.
- **Patrón temporal:** La señal temporal es clara, con un inicio fuerte y una caída suave, lo que ayuda a distinguirla de sonidos transitorios.
