```python
def analizar_audio(filename):
    
    fs, data = wavfile.read(filename) # frecuencia de muestreo y datos de la señal

    # Definición de parámetro temporales
    ts = 1 / fs                     # tiempo de muestreo
    N = len(data)                   # número de muestras en el archivo de audio
    t = np.linspace(0, N * ts, N)   # vector de tiempo
    senial = data#[:, 1]            # se extrae un canal de la pista de audio (si el audio es estereo)
    senial = senial * 3.3 / (2 ** 16 - 1) # se escala la señal a voltios (considerando un CAD de 16bits y Vref 3.3V)
    
    
    #################### Cálculo de la Transformada de Fourier ###################
    
    freq = fft.fftfreq(N, d=1/fs)   # se genera el vector de frecuencias
    senial_fft = fft.fft(senial)    # se calcula la transformada rápida de Fourier
    
    # El espectro es simétrico, nos quedamos solo con el semieje positivo
    f = freq[np.where(freq >= 0)]      
    senial_fft = senial_fft[np.where(freq >= 0)]
    
    # Se calcula la magnitud del espectro
    senial_fft_mod = np.abs(senial_fft) / N     # Respetando la relación de Parseval
    # Al haberse descartado la mitad del espectro, para conservar la energía 
    # original de la señal, se debe multiplicar la mitad restante por dos (excepto
    # en 0 y fm/2)
    senial_fft_mod[1:len(senial_fft_mod-1)] = 2 * senial_fft_mod[1:len(senial_fft_mod-1)]
    if len(senial_fft_mod) > 2:
        senial_fft_mod[1:-1] = 2 * senial_fft_mod[1:-1]
    
    return t, senial, f, senial_fft_mod
```

```python
# Archivos a analizar
archivos = ['alarmas/alarm.wav','alarmas/cooper_fulleon_sounder_tone_3.wav','alarmas/monitor.wav',
            'alarmas/cooper_fulleon_sounder_tone_9.wav','alarmas/cooper_fulleon_sounder_tone_5.wav','alarmas/talk1.wav']

fig, axs = plt.subplots(nrows=6, ncols=2, figsize=(12, 14))  

for i, archivo in enumerate(archivos):
    t, senial, f, senial_fft_mod = analizar_audio(archivo)
    
    # Señal temporal 
    axs[i, 0].plot(t, senial)
    axs[i, 0].set_xlabel('Tiempo [s]')
    axs[i, 0].set_ylabel('Tensión [V]')
    axs[i, 0].set_title(f'Señal temporal: {archivo}')
    axs[i, 0].grid()

    # Espectro de magnitud 
    axs[i, 1].plot(f, senial_fft_mod)
    axs[i, 1].set_xlabel("Frecuencia [Hz]")
    axs[i, 1].set_ylabel("Magnitud")
    axs[i, 1].set_title(f'Espectro de Magnitud: {archivo}')
    axs[i, 1].grid()

plt.tight_layout()
plt.show()
```

## Selección de alarma y justificación

### Requisitos de frecuencia
Para esta aplicación, se requiere que la frecuencia dominante de la alarma esté en el rango de **5 kHz a 8 kHz**. Este rango fue seleccionado porque:
  - Es fácilmente perceptible para la mayoría de las personas.
  - Se encuentra alejado de las frecuencias habituales del ruido ambiental y la voz humana, lo que reduce falsos positivos o confusiones.
  - Facilita el diseño de filtros (por ejemplo, un filtro pasabanda centrado en 6 kHz) para detección automática y robusta, minimizando la interferencia de sonidos externos.

Como se puede observar en los espectros de magnitud de los sonidos analizados ninguno de estos cumplen con la frecuencia necesaria por lo que **se decidió generar una alarma personalizada** con frecuencia en **6 kHz**, asegurando así el cumplimiento de los requisitos técnicos y la eficacia en la detección y diferenciación de la señal.

```python
import numpy as np
from scipy.io.wavfile import write
import IPython

fs = 44100        # Frecuencia de muestreo
f = 6000          # Frecuencia del beep (Hz)
beep_duration = 0.2   # Duración de cada beep en segundos
silence_duration = 0.2 # Duración del silencio entre beeps
n_beeps = 3       # Número de beeps

# Genera un beep de 7kHz
t_beep = np.linspace(0, beep_duration, int(fs * beep_duration), endpoint=False)
beep = 0.5 * np.sin(2 * np.pi * f * t_beep)

# Genera silencio
silence = np.zeros(int(fs * silence_duration))

# Une los beeps y silencios
sound = np.concatenate([np.concatenate([beep, silence]) for _ in range(n_beeps)])

write('beep_6kHz.wav', fs, (sound * 32767).astype(np.int16))
print("¡Archivo beep_6kHz.wav creado!")

IPython.display.Audio('beep_6kHz.wav')
IPython.display.display(IPython.display.FileLink('beep_6kHz.wav'))
```

```python
Audio(filename='alarmas/beep_6kHz.wav')
```

```python
t, senial, f, senial_fft_mod = analizar_audio('alarmas/beep_6kHz.wav')

plt.figure(figsize=(12, 5))

# Gráfica señal temporal
plt.subplot(2, 1, 1)
plt.plot(t, senial)
plt.xlabel('Tiempo [s]')
plt.ylabel('Tensión [V]')
plt.title('Señal temporal: beep_6kHz.wav')
plt.grid()

# Gráfica espectro de magnitud
plt.subplot(2, 1, 2)
plt.plot(f, senial_fft_mod)
plt.xlabel('Frecuencia [Hz]')
plt.xlim(0,10000)
plt.ylabel('Magnitud')
plt.title('Espectro de Magnitud: beep_6kHz.wav')
plt.grid()

plt.tight_layout()
plt.show()
```

## Elección final del sonido. Analisis del espectro de magnitud.
La señal generada, "beep_6kHz", presenta una concentración de energía en la frecuencia de 6 kHz, lo que maximiza su detectabilidad y permite un diseño más eficiente de filtros, a su vez el espectro de magnitud exhibe un único pico definido, minimizando la dispersión energética y reduciendo la probabilidad de confusión con ruidos ambientales o voces humanas. Esta ultima característica es fundamental para asegurar la robustez del sistema de alarma y la eficacia de los algoritmos de detección automática. Por lo que finalmente se elige esta alarma para la aplicación.
